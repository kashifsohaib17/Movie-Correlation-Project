```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\kashi\Downloads\archive\movies.csv')
```

```python
df.head()
```

```python
#searching for and filling missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
df['gross'] = df['gross'].fillna(0)
df['budget'] = df['budget'].fillna(0)
df['votes'] = df['votes'].fillna(0)
df['rating'] = df['rating'].fillna(0)
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#creating a correct year coloumn

df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
#Getting rid of unnecessary decimal points

df['votes'] = df['votes'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['budget'] = df['budget'].astype('int64')
```

```python
df.head()
```

```python
#dropping any duplicates
df.drop_duplicates()
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
#scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Plot Budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color" : "red"}, line_kws={"color": "blue"})
```

```python
#looking closely at the correlation
#df.corr()
df.corr(method = 'pearson')
```

```python
corr_matrix = df.corr(method = 'pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
corr_matrix = df_numerized.corr(method='pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
corr_matrix = df_numerized.corr()

corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
